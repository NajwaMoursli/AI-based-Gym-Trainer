```
PATH = "./gdrive/My Drive/Gym_trainer/Videos_basesDeDonnées/CameraFacePropre"
PATH="."

import os
import numpy as np

files = os.listdir(PATH)
print(files)

file = open(PATH+'Y.txt', "r")
y_train = []
for line in file:
  line = line[:-1]
  line = line.replace(",", ".")
  y_train.append(float(line))

y_train = y_train/(np.max(y_train))
x_train = []

files.sort()
print(files)
#files.pop("Test_RNN.ipynb")
for f in files:
  if f[-3:] == 'npy':
    m = np.load(PATH+f)
    #print(f)
    m = m.reshape(-1)
    x_train.append(m)
    #print(m.shape)
x_train = np.array(x_train)
print((x_train.shape))
#_train[140:].shape
```
